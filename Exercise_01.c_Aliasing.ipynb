## Set up computing environment

```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

## Sample a highly variable signal

We'll sample $\sin(9\cdot x)$ between $0$ and $2\pi$ using 900 and 9 equidistant points.  An we'll see that $\sin(9\cdot x)$ sampled with only 9 points is indistinguishable from $\sin(x)$.

```python
x_hires = np.linspace(0, 2 * np.pi, 900)
x_lores = np.linspace(0, 2 * np.pi, 9)
```

```python
print("delta x at high res", np.diff(x_hires)[0])
print("delta x at low res", np.diff(x_lores)[0])
```

```python
plt.plot(x_hires, np.sin(9 * x_hires), label="sin(9x)")
plt.plot(x_hires, np.sin(x_hires), label="sin(x)")

plt.plot(x_lores, np.sin(9 * x_lores) + 0.05, 'o', label="sin(9x) at low resolution")
plt.plot(x_lores, np.sin(x_lores) - 0.05, 'o', label="sin(x) at low resolution")

plt.gca().xaxis.set_major_locator(plt.MultipleLocator(np.pi / 2))
plt.gca().xaxis.set_minor_locator(plt.MultipleLocator(np.pi / 4))

plt.gca().set_ylim(-1.2, 1.8)

plt.gca().legend(ncol=2)
plt.gca().grid();
```
